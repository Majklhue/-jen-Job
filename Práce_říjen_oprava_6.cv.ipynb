```python
import matplotlib.pyplot as plt
from skimage import io, transform

cesta_k_obrazku = r"C:\Users\Michal\Pictures\TOPPHOTOS\Snímek obrazovky 2024-10-05 231250.png"

try:
    obrazek = io.imread(cesta_k_obrazku)
except FileNotFoundError:
    print(f"Soubor na cestě '{cesta_k_obrazku}' nebyl nalezen!")
    raise

while True:
    try:
        uhel = float(input("Zadej úhel, o který chceš otočit obrázek (v stupních): "))
        break
    except ValueError:
        print("Neplatný vstup! Zadej číslo, například 45 nebo -30.")

otoceny_obrazek = transform.rotate(obrazek, uhel)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Původní obrázek")
plt.imshow(obrazek)
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title(f"Otočený obrázek ({uhel}°)")
plt.imshow(otoceny_obrazek)
plt.axis('off')

plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

cesta = r"C:\Users\Michal\Pictures\Screenshots\Snímek obrazovky 2024-09-22 174830.png"
obrazek_pil = Image.open(cesta)

obrazek = np.array(obrazek_pil)

if obrazek.shape[2] == 4:
    obrazek = cv2.cvtColor(obrazek, cv2.COLOR_RGBA2RGB)

uhel = float(input("Zadej úhel otočení (v stupních): "))

vyska, sirka = obrazek.shape[:2]
stred = (sirka // 2, vyska // 2)
matice = cv2.getRotationMatrix2D(stred, uhel, 1.0)
otoceny = cv2.warpAffine(obrazek, matice, (sirka, vyska))

plt.subplot(1, 2, 1), plt.title("Původní"), plt.imshow(obrazek), plt.axis('off')
plt.subplot(1, 2, 2), plt.title(f"Otočený ({uhel}°)"), plt.imshow(otoceny), plt.axis('off')
plt.show()
```
