```python
import cv2
import numpy as np

cap = cv2.VideoCapture("C:\\Users\\Michal\\Pictures\\WRONG TURN\\TOP SECRET !\\Záznam obrazovky 2024-08-18 230910.mp4")

fps = cap.get(cv2.CAP_PROP_FPS)  
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))  
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))  
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))  

print(f"Počet snímků: {frame_count}")
print(f"FPS: {fps}")
print(f"Rozlišení: {width}x{height}")

# Zpracování videa po snímcích
frame_number = 0
while True:
    ret, frame = cap.read()  # Načte jeden snímek
    if not ret:
        break  

    frame_number += 1
    
    cv2.imshow("Frame", frame)

    # Uvolnění paměti pro snímek
    del frame

    # Stisknu 'q' ukončím zobrazení
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Uvolnění zdrojů
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread("C:\\Users\\Michal\\Pictures\\TOPPHOTOS\\MAJKL a KEVIN.png")

# OpenCV načítá obrázky ve formátu BGR, ale matplotlib zobrazuje RGB, Nejprve převedeš obrázek na RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(image_rgb)
ax[0].set_title("Původní obrázek (RGB)")
ax[0].axis('off')

# Obrázek ve formátu HSV matplotlib nezobrazuje přímo HSV takže se zobrazuje jako RGB, to bohužel není vidět
image_hsv_rgb = cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB)
ax[1].imshow(image_hsv_rgb)
ax[1].set_title("Obrázek v HSV (zobrazený jako RGB)")
ax[1].axis('off')

plt.show()
```
