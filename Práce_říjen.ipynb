```python
import cv2
import numpy as np

cap = cv2.VideoCapture("C:\\Users\\Michal\\Pictures\\WRONG TURN\\TOP SECRET !\\Zznam obrazovky 2024-08-18 230910.mp4")

fps = cap.get(cv2.CAP_PROP_FPS)  
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))  
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))  
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))  

print(f"Poet snmk: {frame_count}")
print(f"FPS: {fps}")
print(f"Rozlien: {width}x{height}")

# Zpracovn videa po snmcch
frame_number = 0
while True:
    ret, frame = cap.read()  # Nate jeden snmek
    if not ret:
        break  

    frame_number += 1
    
    cv2.imshow("Frame", frame)

    # Uvolnn pamti pro snmek
    del frame

    # Stisknu 'q' ukonm zobrazen
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Uvolnn zdroj
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread("C:\\Users\\Michal\\Pictures\\TOPPHOTOS\\MAJKL a KEVIN.png")

# OpenCV nat obrzky ve formtu BGR, ale matplotlib zobrazuje RGB, Nejprve pevede obrzek na RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(image_rgb)
ax[0].set_title("Pvodn obrzek (RGB)")
ax[0].axis('off')

# Obrzek ve formtu HSV matplotlib nezobrazuje pmo HSV take se zobrazuje jako RGB, to bohuel nen vidt
image_hsv_rgb = cv2.cvtColor(image_hsv, cv2.COLOR_HSV2RGB)
ax[1].imshow(image_hsv_rgb)
ax[1].set_title("Obrzek v HSV (zobrazen jako RGB)")
ax[1].axis('off')

plt.show()
```
